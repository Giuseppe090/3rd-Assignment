```python
def longest_increasing_subsequence(arr):
    from bisect import bisect_left

    if not arr:
        return []

    
    lis = []
    prev_indices = [-1] * len(arr)
    positions = [-1] * len(arr)

    for i in range(len(arr)):
        pos = bisect_left(lis, arr[i])
        if pos == len(lis):
            lis.append(arr[i])
        else:
            lis[pos] = arr[i]

        positions[pos] = i
        if pos > 0:
            prev_indices[i] = positions[pos - 1]

    
    lis_length = len(lis)
    lis_sequence = []
    k = positions[lis_length - 1]

    while k != -1:
        lis_sequence.append(arr[k])
        k = prev_indices[k]

    lis_sequence.reverse()
    return lis_sequence

def longest_decreasing_subsequence(arr):
    
    return longest_increasing_subsequence(arr[::-1])[::-1]


def main():
    n = int(input("Inserisci un numero intero positivo n (n ≤ 10000): "))
    if n <= 0 or n > 10000:
        print("Il numero deve essere positivo e non superiore a 10000.")
        return

    perm = list(map(int, input(f"Inserisci una permutazione di lunghezza {n}: ").split()))
    
    if len(perm) != n:
        print("La lunghezza della permutazione non corrisponde a n.")
        return

    lis = longest_increasing_subsequence(perm)
    lds = longest_decreasing_subsequence(perm)

    print("Sottosequenza crescente più lunga:", lis)
    print("Sottosequenza decrescente più lunga:", lds)

if __name__ == "__main__":
    main()
```

```python
def read_fasta(file_path):
    """Legge un file FASTA e restituisce una lista di sequenze."""
    sequences = []
    with open(file_path, 'r') as file:
        sequence = ""
        for line in file:
            line = line.strip()
            if line.startswith(">"):  
                if sequence:  
                    sequences.append(sequence)
                    sequence = ""
            else:
                sequence += line  
        if sequence:  
            sequences.append(sequence)
    return sequences

def find_subsequence_indices(s, t):
    """Trova gli indici in cui i simboli di t appaiono come sottosequenza di s."""
    indices = []
    s_index = 0
    t_index = 0
    
    while s_index < len(s) and t_index < len(t):
        if s[s_index] == t[t_index]:
            indices.append(s_index + 1)  
            t_index += 1
        s_index += 1
    
    return indices if t_index == len(t) else []  

def main():
    
    file_s = 'dna_s.fasta'  
    file_t = 'dna_t.fasta'  

    
    sequences_s = read_fasta(file_s)
    sequences_t = read_fasta(file_t)

    
    if len(sequences_s) != 1 or len(sequences_t) != 1:
        print("Ogni file deve contenere esattamente una sequenza.")
        return

    s = sequences_s[0]
    t = sequences_t[0]

    
    indices = find_subsequence_indices(s, t)

    
    if indices:
        print("Indici in cui i simboli di t appaiono come sottosequenza di s:", indices)
    else:
        print("t non è una sottosequenza di s.")

if __name__ == "__main__":
    main()
```

```python
def longest_common_subsequence(s, t):
    
    m = len(s)
    n = len(t)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s[i - 1] == t[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    
    lcs_length = dp[m][n]
    lcs = []
    i, j = m, n

    while i > 0 and j > 0:
        if s[i - 1] == t[j - 1]:
            lcs.append(s[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1

    lcs.reverse()  
    return ''.join(lcs), lcs_length

def main():
    
    s = "ACGTAG"
    t = "ACGTCG"

    lcs, length = longest_common_subsequence(s, t)
    print(f"La sottosequenza comune più lunga è: {lcs}")
    print(f"Lunghezza della sottosequenza comune più lunga: {length}")

if __name__ == "__main__":
    main()
```

```python
def count_optimal_alignments(s1, s2):
    m = len(s1)
    n = len(s2)

    # memorizzare edit distance
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    # contare allineamenti
    count = [[0] * (n + 1) for _ in range(m + 1)]

    
    for i in range(m + 1):
        dp[i][0] = i  
        count[i][0] = 1  
    for j in range(n + 1):
        dp[0][j] = j  
        count[0][j] = 1  

    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]  
                count[i][j] = count[i - 1][j - 1]  
            else:
                
                options = [
                    (dp[i - 1][j] + 1, count[i - 1][j]),  
                    (dp[i][j - 1] + 1, count[i][j - 1]),  
                    (dp[i - 1][j - 1] + 1, count[i - 1][j - 1])  
                ]
                min_distance = min(option[0] for option in options)
                dp[i][j] = min_distance

                
                for distance, c in options:
                    if distance == min_distance:
                        count[i][j] += c

    return count[m][n]  #allineamenti ottimali

def main():
    # Esempio rosa
    s1 = "PLEASANTLY"
    s2 = "MEANLY"

    num_alignments = count_optimal_alignments(s1, s2)
    print(f"Numero di allineamenti ottimali tra '{s1}' e '{s2}': {num_alignments}")

if __name__ == "__main__":
    main()
```

```python
def alignment_score(seq1, seq2, gap_penalty=-5):
    len1 = len(seq1)
    len2 = len(seq2)

    # memorizzare punteggi
    dp = [[0] * (len2 + 1) for _ in range(len1 + 1)]


    for i in range(len1 + 1):
        dp[i][0] = gap_penalty * i
    for j in range(len2 + 1):
        dp[0][j] = gap_penalty * j

    
    for i in range(1, len1 + 1):
        for j in range(1, len2 + 1):
            match = dp[i - 1][j - 1] + (1 if seq1[i - 1] == seq2[j - 1] else -1)
            delete = dp[i - 1][j] + gap_penalty
            insert = dp[i][j - 1] + gap_penalty
            dp[i][j] = max(match, delete, insert)

    return dp[len1][len2]

# Esempio rosalind
seq1 = "PLEASANTLY"
seq2 = "MEANLY"
score = alignment_score(seq1, seq2)
print("Il punteggio massimo di allineamento è:", score)

#da ripensare con la blosum62
```
